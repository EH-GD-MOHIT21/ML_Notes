```python
##modules import here
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
import numpy as np
import pandas as pd
```

```python
#DataSet
mydib_dataset = datasets.load_diabetes()
print(mydib_dataset.keys())
```

```python
print(mydib_dataset.data)
```

```python
print(mydib_dataset.DESCR)
```

```python
#taking only 1 instance with 1 target
dib_X = mydib_dataset.data[:,np.newaxis,2]
```

```python
dib_X
```

```python
dib_X_train = dib_X[:-30] #last 30
dib_X_test = dib_X[-30:] #start 30
dib_Y_train = mydib_dataset.target[:-30] #last 30
dib_y_test = mydib_dataset.target[-30:] #start 30
```

```python
# creating model with train dataset using linearRegression
model = linear_model.LinearRegression()
model.fit(dib_X_train,dib_Y_train)
dib_y_test_Result = model.predict(dib_X_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
print("Mean squared error is: ",mse(dib_y_test,dib_y_test_Result))
```

```python
print("Weights: ",model.coef_)
print("Intercepts: ",model.intercept_)
```

```python
# Graphical Representation
plt.scatter(dib_X_test, dib_y_test)
plt.plot(dib_X_test, dib_y_test_Result)
```

```python
# taking all the data set and remodeling 
dib_X = mydib_dataset.data
dib_X_train = dib_X[:-30] #last 30
dib_X_test = dib_X[-30:] #start 30
dib_Y_train = mydib_dataset.target[:-30] #last 30
dib_y_test = mydib_dataset.target[-30:] #start 30
# creating model with train dataset using linearRegression
model = linear_model.LinearRegression()
model.fit(dib_X_train,dib_Y_train)
dib_y_test_Result = model.predict(dib_X_test)
from sklearn.metrics import mean_squared_error as mse
print("Mean squared error is: ",mse(dib_y_test,dib_y_test_Result))
print("Weights: ",model.coef_)
print("Intercepts: ",model.intercept_)
```
